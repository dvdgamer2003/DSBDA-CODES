```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from nltk import pos_tag
```

```python
text="This is the sparta. the god of war 3"
token = word_tokenize(text)
```

```python
token
```

```python
p = pos_tag(token)
p
```

```python
stop=set(stopwords.words("english"))
filter=[w for w in token if w.lower() not in stop]
filter
ps=PorterStemmer()
l=WordNetLemmatizer()
a=[ps.stem(w) for w in filter]
b=[l.lemmatize(w) for w in filter]
```
